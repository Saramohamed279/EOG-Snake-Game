```python
import glob
import re
import pandas as pd 
import numpy as np
```

```python
# Dataset Classes:
# ________________

# Yukari-->Up
# Asagi-->Down
# Sag-->Right 
# Sol-->Left
# Kirp-->Blink 
```

```python
# Define the class labels
class_labels = {
    'yukari': 0,
    'yukarı':0 ,
    'yukarih': 0,
    'yukarıh':0 ,
    'yukari': 0,
    'asagi': 1,
    'sag': 2,
    'sol': 3,
    'kirp': 4
}
```

```python
class_names = {
    0: 'Up',
    1: 'Down',
    2: 'Right',
    3: 'Left',
    4: 'Blink'
}
```

```python
 h=[]
classes=class_labels.keys()
```

```python
classes
```

```python
 with open('data handling.csv', 'a') as csv_file:
        for filename in glob.glob('3-class/*.txt'):
            # Get the class label from the filename

            print(filename)
            h = filename.split('\\')[-1].split('.')[0]
            matches = re.split(r'(\d+)', h)
            
            if matches[0] in classes:
                with open(filename, 'r') as file:
                    # Read the contents of the file and split it into lines
                    lines = file.read()
                lines=lines.replace('\n',',')
                for line in lines:
                    csv_file.write(line)
                label = class_labels[matches[0]]
                csv_file.write(str(label)+'\n')
```
